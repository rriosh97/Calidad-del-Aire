```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SparkPandas").getOrCreate()
```

```python
path='/Users/macronald/U Central/Tesis Maestria Calidad/Insumos/Copia de ConsolidadoEstaciones'
dfPandas= pd.read_csv(path, sep=',')
```

```python
dfPandas.head(25)
```

```python
filtro = (dfPandas['DateTime'].str.contains("24:00", na=False)) & (~dfPandas['DateTime'].isnull())
dfPandas.loc[filtro, 'DateTime'] = dfPandas.loc[filtro, 'DateTime'].str.replace("24:00", "23:59") #Validar porque falta agregar un dia cuando es las 00
```

```python
dfPandas.head(25)
```

```python
df= spark.createDataFrame(dfPandas)
```
